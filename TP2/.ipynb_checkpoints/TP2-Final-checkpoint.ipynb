O Fundo de Financiamento ao Estudante do Ensino Superior (FIES) é um programa do Ministério da Educação do Brasil destinado a financiar a graduação na educação superior de estudantes matriculados em instituições não gratuitas. Podem recorrer ao financiamento os estudantes matriculados em cursos superiores que tenham avaliação positiva nos processos conduzidos pelo Ministério da Educação.
   
   Os dados foram selecionados a partir do site de dados públicos do governo federal, na URL¹, onde foram obtidos os dados nos entrega um dicionário de dados que pode ser encontrado na URL², após verificar o dicionário foi percebido que o mesmo não possuia todas a informações dos dados presentes na base, sendo necessário recorer a internet para confimar os mesmos.
   Segue abaixo dados que não se encontram no Dicionário de Dados:
   
    |	SG_AGENTE_FINANCEIRO	|	Sigla do Agente Financeiro	         | 
    |	CO_TIPO_CURSO	        |	Código do Tipo de Curso 	         |
    |	DS_TIPO_CURSO	        |	Descrição do Tipo de Curso 	         |  
    |	CO_PERIODICIDADE_CUR	|	Código da periodicidade do curso	 |
    |	DS_PERIODICIDADE_CUR	|	Descrição da preridiocidade do curso |
    |	NU_ANO_PROC	            |	Número do ano do processo	         | 
    |	NU_MES_PROC	            |	Número do mês do processo	         |
    
   Os dados baixados vem no formato de CSV, foi utilizado a biblioteca do python para ler o arquivo e colocar os dados no formato que o sqlite utiliza para tratar. Por ser uma base de dados muito grande, alguns valores foram removidos do novo esqueme criado, pois não seriam utilizados ou eram redundância, foram removidos CO_CONTRATO_FIES, SG_AGENTE_FINANCEIRO, NU_MES, NU_SEMESTRE, NU_ANO, VL_MENSALIDADE, NU_ANO_EXERCICIO_INSC, DS_RACA_COR, NO_IES_EXT_ALUNOS, NU_ANO_PROC, NU_MES_PROC.
   A as próximas células são onde ocorre o processo de tratamento dos dados para o formato do sqlite.

Criar banco de dados

```python
import sqlite3 as sl
import pandas as pd
import csv

con = sl.connect("dados_fies.db")
cur = con.cursor()
cur.executescript("PRAGMA foreign_keys=0")
banco_de_dados = """
BEGIN TRANSACTION;
CREATE TABLE IF NOT EXISTS "Estudante" (
	"CO_INSCRICAO_EXT_ALUNOS"	INTEGER,
	"CO_CONTRATO_FIES_EXT_ALUNOS"	INTEGER,
	"CO_IES_EXT_ALUNOS"	INTEGER,
	"NU_PERCENTUAL_PROUNI"	REAL,
	"NU_PERCENT_SOLICITADO_FINANC"	REAL,
	"DT_NASCIMENTO"	DATE,
	"ST_DEFICIENCIA"	TEXT,
	"ST_ENSINO_MEDIO_ESCOLA_PUBLICA"	TEXT,
	"CO_CIDADE"	INTEGER,
	"SG_SEXO"	TEXT,
	"DS_SEXO"	TEXT,
	"CO_ESTADO_CIVIL"	INTEGER,
	"DS_ESTADO_CIVIL"	TEXT,
	"SG_RACA_COR"	TEXT,
	"VL_MENSALIDADE_EXT_ALUNOS"	REAL,
	"CO_ADITAMENTO_EXT_ALUNOS"	INTEGER,
    PRIMARY KEY("CO_INSCRICAO_EXT_ALUNOS","CO_CONTRATO_FIES_EXT_ALUNOS","CO_ADITAMENTO_EXT_ALUNOS"),
	FOREIGN KEY("CO_CIDADE") REFERENCES "Municipio"("CO_MUNICIPIO"),
	FOREIGN KEY("CO_IES_EXT_ALUNOS") REFERENCES "IES"("CO_IES")
);
CREATE TABLE IF NOT EXISTS "Mantenedoura" (
	"CO_MANTENEDORA"	INTEGER,
	"NU_CNPJ_MANTENEDORA"	INTEGER,
	"NO_RAZAO_SOCIAL_MANTENEDORA"	TEXT,
	"CO_MUNICIPIO_MANTENEDORA"	INTEGER,
	PRIMARY KEY("CO_MANTENEDORA","NU_CNPJ_MANTENEDORA"),
	FOREIGN KEY("CO_MUNICIPIO_MANTENEDORA") REFERENCES "Municipio"("CO_MUNICIPIO")
);
CREATE TABLE IF NOT EXISTS "Curso" (
	"DS_CURSO"	INTEGER ,
	"CO_CURSO"	INTEGER ,
	"CO_IES"	INTEGER ,
    "CO_TIPO_CURSO" INTEGER,
    "DS_TIPO_CURSO" TEXT,
    "CO_PERIODICIDADE_CUR" INTEGER,
    "DS_PERIODICIDADE_CUR" TEXT,
    PRIMARY KEY("CO_CURSO", "CO_IES"),
	FOREIGN KEY("CO_IES") REFERENCES "IES"("CO_IES")
);
CREATE TABLE IF NOT EXISTS "IES" (
	"CO_IES"	INTEGER ,
	"CO_CAMPUS"	INTEGER ,
    "NO_IES"	TEXT ,
	"CO_MUNICIPIO_IES"	INTEGER ,
	"NO_CAMPUS"	TEXT ,
	PRIMARY KEY("CO_IES","CO_CAMPUS"),
	FOREIGN KEY("CO_MUNICIPIO_IES") REFERENCES "Municipio"("CO_MUNICIPIO")
);
CREATE TABLE IF NOT EXISTS "Processo" (
	"CO_PROCESSO"	INTEGER,
	"CO_INSCRICAO"	INTEGER,
    "CO_ADITAMENTO" INTEGER,
 	"CO_MANTENEDORA"	INTEGER,
	"CO_CURSO"	INTEGER,
	"CO_AGENTE_FINANCEIRO"	INTEGER,
	"NO_AGENTE_FINANCEIRO"	TEXT,
	"VL_SEMESTRE"	REAL,
	"TP_FIANCA"	TEXT,
	"VL_REPASSE"	REAL,
	"VL_PERC_FINANCIAMENTO"	REAL,
	"QT_SEMESTRE_FINANCIADO"	INTEGER,
    "NU_ANO" INTEGER,
	PRIMARY KEY("CO_PROCESSO","CO_INSCRICAO","CO_ADITAMENTO"),
	FOREIGN KEY("CO_MANTENEDORA") REFERENCES "Mantenedoura"("CO_MANTENEDORA"),
	FOREIGN KEY("CO_CURSO") REFERENCES "Curso"("CO_CURSO"),
    FOREIGN KEY("CO_INSCRICAO") REFERENCES "Estudante"("CO_INSCRICAO_EXT_ALUNOS")
);
CREATE TABLE IF NOT EXISTS "Municipio" (
	"SG_UF"	TEXT,
	"NO_MUNICIPIO"	TEXT,
    "CO_MUNICIPIO"	INTEGER,
    PRIMARY KEY("CO_MUNICIPIO")
    
);
COMMIT;
"""
cur.executescript(banco_de_dados)
print("Banco de dados criado com sucesso")
```

```python
#1 mostrar os nomes de todos os cursos de Licenciatura financiados pelo FIES
sqlquery="""
select distinct DS_CURSO from Curso 
where CO_CURSO IN (
select distinct CO_CURSO from Processo) and DS_TIPO_CURSO='Licenciatura'
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#1 mostrar os nomes de todos os cursos de Licenciatura financiados pelo FIES
sqlquery="""
select distinct DS_CURSO from Curso natural join Processo
where DS_TIPO_CURSO='Licenciatura'
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#2 mostrar os códigos de contrato de todos os estudantes do FIES com alguma deficiência
sqlquery="""
select CO_CONTRATO_FIES_EXT_ALUNOS from Estudante 
where ST_DEFICIENCIA='S'
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#3 mostrar todos os processos feitos para estudantes casados
sqlquery="""
select * from Processo 
where CO_INSCRICAO in (
select CO_INSCRICAO_EXT_ALUNOS from Estudante
where CO_ESTADO_CIVIL=2
)
order by CO_INSCRICAO
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#3, mostrar todos os processos feitos para estudantes casados, só que com menos colunas porque não tinha necessidade de copiar tudo
sqlquery="""
select distinct CO_INSCRICAO, CO_PROCESSO, CO_ADITAMENTO from
(select * from Estudante join Processo where CO_INSCRICAO=CO_INSCRICAO_EXT_ALUNOS)
where CO_ESTADO_CIVIL=2
order by CO_INSCRICAO
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#4  -- mostrar todas as faculdades com cursos de Sistemas de Informação
sqlquery="""
select distinct NO_IES from IES 
where CO_IES in
(select CO_IES from Curso 
where DS_CURSO='SISTEMAS DE INFORMAÇÃO' or DS_CURSO='SISTEMA DE INFORMAÇÃO')
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#4  -- mostrar todas as faculdades com cursos de Sistemas de Informação
sqlquery="""
select distinct NO_IES from 
(IES natural join Curso)
where DS_CURSO='SISTEMAS DE INFORMAÇÃO' or DS_CURSO='SISTEMA DE INFORMAÇÃO'
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#5 -- mostrar os números de contrato de todos os estudantes com financiamento maior que 50%
sqlquery="""
select e.CO_CONTRATO_FIES_EXT_ALUNOS as cod_contrato
from Estudante e
join Processo p on p.CO_INSCRICAO = e.CO_INSCRICAO_EXT_ALUNOS
where p.VL_PERC_FINANCIAMENTO > 50
group by p.CO_INSCRICAO;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#5 -- mostrar os números de contrato de todos os estudantes com financiamento maior que 50%
sqlquery="""
select e.CO_CONTRATO_FIES_EXT_ALUNOS as cod_contrato
from Estudante e, Processo p
where p.VL_PERC_FINANCIAMENTO > 50
	and p.CO_INSCRICAO = e.CO_INSCRICAO_EXT_ALUNOS
group by p.CO_INSCRICAO;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#    -- (6) mostrar os codigos do contrato de todos os alunos financiando cursos de Sistema de Informação
sqlquery="""
select CO_INSCRICAO_EXT_ALUNOS as contrato_aluno
    from Estudante e
    join Processo p on e.CO_INSCRICAO_EXT_ALUNOS = p.CO_INSCRICAO
    join Curso c on c.CO_CURSO = p.CO_CURSO
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    group by contrato_aluno;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#    -- (6) mostrar os codigos do contrato de todos os alunos financiando cursos de Sistema de Informação
sqlquery="""
select CO_INSCRICAO_EXT_ALUNOS as contrato_aluno
    from Estudante e, Processo p, Curso c
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
        and e.CO_INSCRICAO_EXT_ALUNOS = p.CO_INSCRICAO
        and c.CO_CURSO = p.CO_CURSO
    group by contrato_aluno;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (7) mostrar os nomes de todas as mantenedoras que financiam cursos de Sistemas de Informação
sqlquery="""
    select NO_RAZAO_SOCIAL_MANTENEDORA as financiadora
    from Mantenedoura m
    join Processo p on m.CO_MANTENEDORA = p.CO_MANTENEDORA
    join Curso c on c.CO_CURSO = p.CO_CURSO
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    group by financiadora;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (7) mostrar os nomes de todas as mantenedoras que financiam cursos de Sistemas de Informação
sqlquery="""
    select NO_RAZAO_SOCIAL_MANTENEDORA as financiadora
    from Mantenedoura m, Processo p, Curso c
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    	and m.CO_MANTENEDORA = p.CO_MANTENEDORA
    	and c.CO_CURSO = p.CO_CURSO
    group by financiadora;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (8) mostrar os nomes de todos os municípios com cursos de Sistemas de Informação
sqlquery="""
   select NO_MUNICIPIO as nome_municipio
    from Municipio m
    join IES i on m.CO_MUNICIPIO = i.CO_MUNICIPIO_IES
    join Curso c on c.CO_IES = i.CO_IES
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    group by nome_municipio;

"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (8) mostrar os nomes de todos os municípios com cursos de Sistemas de Informação
sqlquery="""
    select NO_MUNICIPIO as nome_municipio
    from Municipio m, IES i, Curso c
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
        and m.CO_MUNICIPIO = i.CO_MUNICIPIO_IES
        and c.CO_IES = i.CO_IES
    group by nome_municipio;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (9) mostrar o valor total dos financiamentos para cursos de Sistemas de Informação
sqlquery="""
select sum(valor_investido) from (
    	select sum(distinct VL_REPASSE) as valor_investido
    	from Processo p
    	join Curso c on c.CO_CURSO = p.CO_CURSO
    	where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    	group by p.CO_INSCRICAO);
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
#-- (9) mostrar o valor total dos financiamentos para cursos de Sistemas de Informação
sqlquery="""
    select sum(valor_investido) from (
    	select sum(distinct VL_REPASSE) as valor_investido
    	from Processo p, Curso c
    	where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    		and c.CO_CURSO = p.CO_CURSO
    	group by p.CO_INSCRICAO);
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
 #   -- (10) quantidade de estudantes financiando 100% do curso de Sistemas de Informação
sqlquery="""
   SELECT count(DISTINCT CO_INSCRICAO) as qtd_bolsistas_sistemas_informacao
    from Processo p
    join Curso c on c.CO_CURSO = p.CO_CURSO
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    	and p.VL_PERC_FINANCIAMENTO = 100;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```

```python
 #   -- (10) quantidade de estudantes financiando 100% do curso de Sistemas de Informação
sqlquery="""
    SELECT count(DISTINCT CO_INSCRICAO) as qtd_bolsistas_sistemas_informacao
    from Processo p, Curso c
    where c.DS_CURSO = 'SISTEMAS DE INFORMAÇÃO'
    	and p.VL_PERC_FINANCIAMENTO = 100
        and c.CO_CURSO = p.CO_CURSO;
"""
dfq1 = pd.read_sql_query(sqlquery,con)
dfq1
```
