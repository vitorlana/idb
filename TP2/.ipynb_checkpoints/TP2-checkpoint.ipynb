```python
import sqlite3 as sl
import pandas as pd
import csv

con = sl.connect("dados_fies.db")
cur = con.cursor()
cur.executescript("PRAGMA foreign_keys=0")
banco_de_dados = """
BEGIN TRANSACTION;
DROP TABLE IF EXISTS "Estudante";
CREATE TABLE IF NOT EXISTS "Estudante" (
	"CO_INSCRICAO_EXT_ALUNOS"	INTEGER,
	"CO_CONTRATO_FIES_EXT_ALUNOS"	INTEGER,
	"CO_IES_EXT_ALUNOS"	INTEGER,
	"NU_PERCENTUAL_PROUNI"	REAL,
	"NU_PERCENT_SOLICITADO_FINANC"	REAL,
	"DT_NASCIMENTO"	DATE,
	"ST_DEFICIENCIA"	TEXT,
	"ST_ENSINO_MEDIO_ESCOLA_PUBLICA"	TEXT,
	"CO_CIDADE"	INTEGER,
	"SG_SEXO"	TEXT,
	"DS_SEXO"	TEXT,
	"CO_ESTADO_CIVIL"	INTEGER,
	"DS_ESTADO_CIVIL"	TEXT,
	"SG_RACA_COR"	TEXT,
	"VL_MENSALIDADE_EXT_ALUNOS"	REAL,
	"CO_ADITAMENTO_EXT_ALUNOS"	INTEGER,
	--"NU_ANO_PROC" INTEGER,
    --"NU_MES_PROC" INTEGER,
    PRIMARY KEY("CO_INSCRICAO_EXT_ALUNOS","CO_CONTRATO_FIES_EXT_ALUNOS","CO_ADITAMENTO_EXT_ALUNOS"),
	FOREIGN KEY("CO_CIDADE") REFERENCES "Municipio"("CO_MUNICIPIO"),
	FOREIGN KEY("CO_IES_EXT_ALUNOS") REFERENCES "IES"("CO_IES")
);
DROP TABLE IF EXISTS "Mantenedoura";
CREATE TABLE IF NOT EXISTS "Mantenedoura" (
	"CO_MANTENEDORA"	INTEGER,
	"NU_CNPJ_MANTENEDORA"	INTEGER,
	"NO_RAZAO_SOCIAL_MANTENEDORA"	TEXT,
	"CO_MUNICIPIO_MANTENEDORA"	INTEGER,
	PRIMARY KEY("CO_MANTENEDORA","NU_CNPJ_MANTENEDORA"),
	FOREIGN KEY("CO_MUNICIPIO_MANTENEDORA") REFERENCES "Municipio"("CO_MUNICIPIO")
);
DROP TABLE IF EXISTS "Curso";
CREATE TABLE IF NOT EXISTS "Curso" (
	"DS_CURSO"	INTEGER ,
	"CO_CURSO"	INTEGER ,
	"CO_IES"	INTEGER ,
    "CO_TIPO_CURSO" INTEGER,
    "DS_TIPO_CURSO" TEXT,
    "CO_PERIODICIDADE_CUR" INTEGER,
    "DS_PERIODICIDADE_CUR" TEXT,
    PRIMARY KEY("CO_CURSO", "CO_IES"),
	FOREIGN KEY("CO_IES") REFERENCES "IES"("CO_IES")
);
DROP TABLE IF EXISTS "IES";
CREATE TABLE IF NOT EXISTS "IES" (
	"CO_IES"	INTEGER ,
	"CO_CAMPUS"	INTEGER ,
    "NO_IES"	TEXT ,
	"CO_MUNICIPIO_IES"	INTEGER ,
	"NO_CAMPUS"	TEXT ,
	PRIMARY KEY("CO_IES","CO_CAMPUS"),
	FOREIGN KEY("CO_MUNICIPIO_IES") REFERENCES "Municipio"("CO_MUNICIPIO")
);
DROP TABLE IF EXISTS "Processo";
CREATE TABLE IF NOT EXISTS "Processo" (
	"CO_PROCESSO"	INTEGER,
	"CO_INSCRICAO"	INTEGER,
    "CO_ADITAMENTO" INTEGER,
 	"CO_MANTENEDORA"	INTEGER,
	"CO_CURSO"	INTEGER,
	"CO_AGENTE_FINANCEIRO"	INTEGER,
	"NO_AGENTE_FINANCEIRO"	TEXT,
	"VL_SEMESTRE"	REAL,
	"TP_FIANCA"	TEXT,
	"VL_REPASSE"	REAL,
	"VL_PERC_FINANCIAMENTO"	REAL,
	"QT_SEMESTRE_FINANCIADO"	INTEGER,
    "NU_ANO" INTEGER,
	PRIMARY KEY("CO_PROCESSO","CO_INSCRICAO","CO_ADITAMENTO"),
	FOREIGN KEY("CO_MANTENEDORA") REFERENCES "Mantenedoura"("CO_MANTENEDORA"),
	FOREIGN KEY("CO_CURSO") REFERENCES "Curso"("CO_CURSO"),
    FOREIGN KEY("CO_INSCRICAO") REFERENCES "Estudante"("CO_INSCRICAO_EXT_ALUNOS")
);
DROP TABLE IF EXISTS "Municipio";
CREATE TABLE IF NOT EXISTS "Municipio" (
	"SG_UF"	TEXT,
	"NO_MUNICIPIO"	TEXT,
    "CO_MUNICIPIO"	INTEGER,
    PRIMARY KEY("CO_MUNICIPIO")
    
);
COMMIT;
"""
cur.executescript(banco_de_dados)
print("Banco de dados criado com sucesso")
```
